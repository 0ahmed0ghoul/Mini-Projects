```python
#Matplotlib (matplotlib) :

#A foundational library for creating static, animated, and interactive visualizations.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a plot
plt.plot(x, y, label="sin(x)")
plt.title("Sine Wave")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Line Plot
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label="sin(x)", color="blue", linestyle="-")
plt.title("Line Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Scatter Plot
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap="viridis")
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.colorbar()
plt.show()
```

```python
# Bar Chart
```

```python
categories = ["A", "B", "C", "D"]
values = [10, 20, 15, 25]

plt.bar(categories, values, color="skyblue")
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
# Horizontal Bar Chart
```

```python
plt.barh(categories, values, color="lightgreen")
plt.title("Horizontal Bar Chart")
plt.xlabel("Values")
plt.ylabel("Categories")
plt.show()
```

```python
# Histogram
```

```python
data = np.random.normal(0, 1, 1000)

plt.hist(data, bins=30, color="orange", edgecolor="black")
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Box Plot
```

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

plt.boxplot(data, patch_artist=True, boxprops=dict(facecolor="lightblue"))
plt.title("Box Plot")
plt.xlabel("Dataset")
plt.ylabel("Value")
plt.show()
```

```python
# Pie Chart
```

```python
labels = ["A", "B", "C", "D"]
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # "Explode" the 2nd slice

plt.pie(sizes, explode=explode, labels=labels, autopct="%1.1f%%", startangle=90)
plt.title("Pie Chart")
plt.show()
```

```python
# Image Plot
```

```python
image = np.random.rand(10, 10)

plt.imshow(image, cmap="hot")
plt.title("Image Plot")
plt.colorbar()
plt.show()
```

```python
# Contour Plot
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

plt.contour(X, Y, Z, levels=20, cmap="RdGy")
plt.title("Contour Plot")
plt.colorbar()
plt.show()
```

```python
# Filled Contour Plot
```

```python
plt.contourf(X, Y, Z, levels=20, cmap="RdGy")
plt.title("Filled Contour Plot")
plt.colorbar()
plt.show()
```

```python
# Quiver Plot (Vector Field
```

```python
x = np.linspace(-2, 2, 10)
y = np.linspace(-2, 2, 10)
X, Y = np.meshgrid(x, y)
U = np.cos(X)
V = np.sin(Y)

plt.quiver(X, Y, U, V)
plt.title("Quiver Plot")
plt.show()
```

```python
# Stem Plot
```

```python
x = np.linspace(0, 10, 20)
y = np.cos(x)

plt.stem(x, y, linefmt="b-", markerfmt="bo", basefmt="r-")
plt.title("Stem Plot")
plt.show()
```

```python
# Step Plot
```

```python
plt.step(x, y, where="mid", label="Step Plot")
plt.title("Step Plot")
plt.legend()
plt.show()
```

```python
# Error Bar Plot
```

```python
x = np.arange(5)
y = [1, 3, 2, 5, 4]
y_err = [0.1, 0.3, 0.2, 0.5, 0.4]

plt.errorbar(x, y, yerr=y_err, fmt="o", capsize=5)
plt.title("Error Bar Plot")
plt.show()
```

```python
# Fill Between
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.fill_between(x, y1, y2, color="lightblue", alpha=0.5)
plt.title("Fill Between")
plt.show()
```

```python
# Pcolormesh (Pseudocolor Plot)
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

plt.pcolormesh(X, Y, Z, cmap="viridis")
plt.title("Pcolormesh")
plt.colorbar()
plt.show()
```

```python
# Streamplot
```

```python
x = np.linspace(-2, 2, 10)
y = np.linspace(-2, 2, 10)
X, Y = np.meshgrid(x, y)
U = np.cos(X)
V = np.sin(Y)

plt.streamplot(X, Y, U, V, density=1.5, color=U, linewidth=2, cmap="autumn")
plt.title("Streamplot")
plt.show()
```

```python
# 3D Line Plot
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

z = np.linspace(0, 15, 100)
x = np.sin(z)
y = np.cos(z)

ax.plot3D(x, y, z, "blue")
ax.set_title("3D Line Plot")
plt.show()
```

```python
# 3D Scatter Plot
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

ax.scatter3D(x, y, z, c=z, cmap="viridis")
ax.set_title("3D Scatter Plot")
plt.show()
```

```python
# 3D Surface Plot
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap="viridis")
ax.set_title("3D Surface Plot")
plt.show()
```
