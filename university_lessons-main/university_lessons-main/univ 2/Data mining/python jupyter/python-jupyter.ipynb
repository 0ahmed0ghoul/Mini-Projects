```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt  # Use plt for consistency

# Read the CSV file
df = pd.read_csv(r"C:\Users\user\Desktop\Data mining\heart[1].csv")

# Compute the correlation matrix from the DataFrame
corr_matrix = df.corr()

# Create a figure for the heatmap
plt.figure(figsize=(10, 8))

# Plot the heatmap of the correlation matrix
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=10)

# Add a title to the plot
plt.title("Correlation Matrix Heatmap")

# Show the plot
plt.show()
```
