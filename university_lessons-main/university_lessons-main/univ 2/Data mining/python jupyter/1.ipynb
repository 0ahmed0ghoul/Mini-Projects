```python
from scipy.stats import zscore
z_score = np.abs(zscore(df))
print('z_score:')
z_score

outliers = (z_score>3)
print('outliers based on z-score >3 :')
print(np.sum(outliers))
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3-Q1
outliers = (df<(Q1-1.5*IQR))|(df>(Q3+1.5*IQR))
np.sum(outliers)
```
